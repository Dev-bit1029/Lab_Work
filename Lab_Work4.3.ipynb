```python
# Create an ID array (list)
id_list = [101, 102, 103, 104, 105]

# Display original list
print("Original ID Array:", id_list)

# Take input from user
value = int(input("Enter the value to delete: "))

# Check and delete the element
if value in id_list:
    id_list.remove(value)
    print("Element deleted successfully.")
else:
    print("Value not found in the list.")

# Display updated list
print("Updated ID Array:", id_list)
```

```python
# Create an ID array (list)
id_list = [101, 102, 103, 104, 105]

# Display original list
print("Original ID Array:", id_list)

# Take input from user
index = int(input("Enter index to update (0 to n-1): "))
new_value = int(input("Enter new value: "))

# Check if index is valid
if 0 <= index < len(id_list):
    id_list[index] = new_value
    print("Element updated successfully.")
else:
    print("Invalid index!")

# Display updated list
print("Updated ID Array:", id_list)
```

```python
# Create an ID array (list)
id_list = [101, 102, 103, 104, 105]

# Display list
print("ID Array:", id_list)

# Take input from user
value = int(input("Enter element to search: "))

# Search for the element
if value in id_list:
    index = id_list.index(value)
    print("Element found at index:", index)
else:
    print("Element not found in the array.")
```
