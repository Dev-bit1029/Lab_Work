```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact

num = int(input("Enter a number: "))

print("Factorial is:", factorial(num))
```

```python

def cube(x):
    return x*x*x

def show(nums, func):
    for i in nums:
        print(func(i))

numbers = [1, 2, 3, 4]

show(numbers, cube)
```

```python
def calculate(*nums):
    total = 0
    product = 1
    
    for n in nums:
        total += n
        product *= n
    
    return total, product

# call function
s, p = calculate(1, 2, 3, 4)

print("Sum:", s)
print("Product:", p)
```

```python
def calc(*n):
    s = sum(n)
    p = 1
    for i in n:
        p *= i
    return s, p

print(calc(1, 2))
```

```python
def show(*names):
    if not names:
        print("No students")
    else:
        for n in names:
            print(n)

# calls
show("A", "B", "C")
show()
```

```python
def separate(*args):
    strings = ()
    numbers = ()
    
    for x in args:
        if type(x) == str:
            strings = strings + (x,)
        else:
            numbers = numbers + (x,)
    
    return strings, numbers

# example
s, n = separate(1, "A", 2, "B", 3, "C")

print("Strings:", s)
print("Numbers:", n)
```

```python
def area(l, w):
    """Returns area of rectangle"""
    return l * w

print(area(5, 3))        # area
print(area.__doc__)      # docstring
```

```python
def fib(n):
    """Returns Fibonacci numbers up to n"""
    a, b = 0, 1
    res = []
    
    while a <= n:
        res.append(a)
        a, b = b, a + b
    
    return res

print(fib(10))
```
