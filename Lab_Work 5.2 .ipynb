```python
class Demo:
    def __init__(self, name):
        self.name = name
        print(f"Object {self.name} created")

    def __del__(self):
        print(f"Object {self.name} destroyed")


# Creating multiple objects
obj1 = Demo("A")
obj2 = Demo("B")
obj3 = Demo("C")

print("\nNow deleting objects...\n")

# Deleting objects using del keyword
del obj1
del obj2
del obj3

print("End of program")
```

```python
class Animal:
    def __init__(self, name):
        self.name = name  # initializing name attribute

    def display(self):
        print("Animal Name:", self.name)


# Creating objects of the class
a1 = Animal("Dog")
a2 = Animal("Cat")

# Displaying animal names
a1.display()
a2.display()
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        area = self.length * self.width
        print("Area of Rectangle:", area)


# Creating objects with parameters
r1 = Rectangle(10, 5)
r2 = Rectangle(7, 3)

# Calculating area
r1.calculate_area()
r2.calculate_area()
```

```python
class Employee:
    # Default constructor
    def __init__(self):
        self.name = "John"
        self.id = 101
        print("Employee object created")

    # Destructor
    def __del__(self):
        print(f"Goodbye {self.name}, Employee object is being deleted")


# Creating object
e1 = Employee()

# Deleting object
del e1

print("Program End")
```

```python
class Student:
    # Constructor to initialize attributes
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    # Method to display student details
    def display(self):
        print("Student Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)

    # Method to calculate result
    def result(self):
        if self.marks >= 40:
            print("Result: Pass")
        else:
            print("Result: Fail")


# Creating objects
s1 = Student("Amit", 101, 75)
s2 = Student("Neha", 102, 35)

# Displaying details and result
s1.display()
s1.result()

print()

s2.display()
s2.result()
```
