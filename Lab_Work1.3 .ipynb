```python

a = 10                     
b = 3.14                   
c = "Hello"                
d = True                   
e = [1, 2, 3]              
f = (4, 5, 6)              
g = {"name": "Dev", "age": 20}   

print("Integer:", a, type(a), id(a))
print("Float:", b, type(b), id(b))
print("String:", c, type(c), id(c))
print("Boolean:", d, type(d), id(d))
print("List:", e, type(e), id(e))
print("Tuple:", f, type(f), id(f))
print("Dictionary:", g, type(g), id(g))
```
