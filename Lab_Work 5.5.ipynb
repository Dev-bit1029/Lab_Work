```python
from abc import ABC, abstractmethod
import math

# Abstract class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


# Rectangle class implementing both methods
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


# Circle class implementing both methods
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


# Correct usage
rect = Rectangle(10, 5)
circle = Circle(7)

print("Rectangle Area:", rect.area())
print("Rectangle Perimeter:", rect.perimeter())

print("\nCircle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())


# ❌ Incorrect subclass (missing perimeter implementation)
class IncompleteShape(Shape):
    def area(self):
        return 100


# Try instantiating incomplete class
try:
    obj = IncompleteShape()
except TypeError as e:
    print("\nError:", e)
```

```python
from abc import ABC, abstractmethod

# Abstract class
class MLModel(ABC):

    @abstractmethod
    def train(self):
        pass

    @abstractmethod
    def predict(self):
        pass


# Subclass 1: Linear Regression Model
class LinearRegressionModel(MLModel):

    def train(self):
        print("Training Linear Regression Model using least squares method...")

    def predict(self):
        print("Predicting using Linear Regression equation: y = mx + c")


# Subclass 2: Decision Tree Model
class DecisionTreeModel(MLModel):

    def train(self):
        print("Training Decision Tree Model using entropy / gini split...")

    def predict(self):
        print("Predicting using Decision Tree rules and conditions...")


# Creating objects
lr_model = LinearRegressionModel()
dt_model = DecisionTreeModel()

# Same interface, different implementations
print("=== Linear Regression Model ===")
lr_model.train()
lr_model.predict()

print("\n=== Decision Tree Model ===")
dt_model.train()
dt_model.predict()
```

```python
from abc import ABC, abstractmethod

# =========================
# ABSTRACTION
# =========================
class Account(ABC):

    @abstractmethod
    def deposit(self, amount):
        pass

    @abstractmethod
    def withdraw(self, amount):
        pass


# =========================
# ENCAPSULATION + BASE CLASS
# =========================
class BankAccount(Account):

    def __init__(self, account_number, balance=0):
        self.__account_number = account_number   # private attribute
        self.__balance = balance                # private attribute

    # Encapsulated method
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")

    # Basic withdraw (can be overridden)
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number


# =========================
# INHERITANCE
# Savings Account
# =========================
class SavingsAccount(BankAccount):

    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.get_balance() * self.interest_rate / 100
        self.deposit(interest)
        print(f"Interest added: {interest}")


# =========================
# POLYMORPHISM (Method Overriding)
# Current Account
# =========================
class CurrentAccount(BankAccount):

    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    # overridden withdraw method
    def withdraw(self, amount):
        available = self.get_balance() + self.overdraft_limit

        if amount <= available:
            # manually adjust balance via deposit/withdraw logic
            if amount <= self.get_balance():
                super().withdraw(amount)
            else:
                overdraft_used = amount - self.get_balance()
                print(f"Using overdraft: {overdraft_used}")
                # set balance to 0 after overdraft usage
                self._BankAccount__balance = 0
        else:
            print("Overdraft limit exceeded")


# =========================
# TEST SECTION
# =========================

print("=== Savings Account ===")
s1 = SavingsAccount("SA101", 1000, 5)
s1.deposit(500)
s1.add_interest()
s1.withdraw(200)
print("Balance:", s1.get_balance())

print("\n=== Current Account ===")
c1 = CurrentAccount("CA202", 1000, 500)
c1.deposit(300)
c1.withdraw(1200)   # uses overdraft
c1.withdraw(300)
print("Balance:", c1.get_balance())
```
