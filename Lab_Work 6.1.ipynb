```python
def analyze_file(filename):
    try:
        with open(filename, "r") as file:
            lines = file.readlines()
            line_count = len(lines)
            char_count = 0
            word_count = 0
            
            for line in lines:
                char_count += len(line)
                word_count += len(line.split())
                
            print(f"Lines: {line_count}")
            print(f"Words: {word_count}")
            print(f"Characters: {char_count}")
    except FileNotFoundError:
        print("File not found.")

analyze_file("notes.txt")
```

```python
search_word = input("Enter word to search: ").lower()

try:
    with open("sample.txt", "r") as file:
        found = False
        for line_num, line in enumerate(file, 1):
            if search_word in line.lower():
                print(f"Found '{search_word}' at Line {line_num}")
                found = True
        if not found:
            print("Word not found.")
except FileNotFoundError:
    print("sample.txt not found.")
```

```python
try:
    with open("source.txt", "r") as src, open("backup.txt", "w") as dest:
        dest.write(src.read())
    print("File copied successfully from source.txt to backup.txt.")
except FileNotFoundError:
    print("Source file not found.")
```

```python
modes = ['r', 'w', 'a', 'r+', 'w+', 'a+']
filename = "demo.txt"

# Ensure file exists for 'r' and 'r+' demos
with open(filename, "w") as f: f.write("Initial data.")

for mode in modes:
    print(f"\n--- Testing Mode: {mode} ---")
    try:
        with open(filename, mode) as f:
            if 'r' in mode:
                # Need to seek to start if mode is r+ or w+ after writing
                if '+' in mode: f.seek(0)
                print(f"Read operation: {f.read()}")
            
            if 'w' in mode or 'a' in mode or '+' in mode:
                f.write(f"\nWritten in {mode} mode.")
                print(f"Write/Append operation performed.")
    except Exception as e:
        print(f"Error in {mode}: {e}")
```
