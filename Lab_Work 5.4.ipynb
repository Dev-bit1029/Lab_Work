```python
def add(a, b):
    # If both are integers, perform addition
    if isinstance(a, int) and isinstance(b, int):
        return a + b
    
    # If both are strings, perform concatenation
    elif isinstance(a, str) and isinstance(b, str):
        return a + b
    
    else:
        return "Invalid input: both arguments must be of same type (int or str)"


# Testing the function
print("Adding integers:", add(10, 20))
print("Concatenating strings:", add("Hello, ", "World!"))
print("Invalid case:", add(10, "Hello"))
```

```python
class Shape:
    def area(self):
        print("Area method of Shape class (generic)")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Creating objects
s1 = Circle(5)
s2 = Rectangle(4, 6)

# Polymorphism in action
shapes = [s1, s2]

for shape in shapes:
    print("Area:", shape.area())
```

```python
# Interface-like base class
class Transport:
    def travel(self):
        pass  # This method will be overridden


# Derived class 1
class Train(Transport):
    def travel(self):
        print("Train travels on tracks at medium speed.")


# Derived class 2
class Plane(Transport):
    def travel(self):
        print("Plane travels in the sky at high speed.")


# Creating objects
t1 = Train()
t2 = Plane()

# Polymorphism in action
vehicles = [t1, t2]

for vehicle in vehicles:
    vehicle.travel()
```

```python
class Calculator:
    def multiply(self, *args):
        result = 1
        
        # Multiply all arguments passed
        for num in args:
            result *= num
        
        return result


# Creating object
calc = Calculator()

# Calling with 2 arguments
print("Multiplication of 2 numbers:", calc.multiply(5, 10))

# Calling with 3 arguments
print("Multiplication of 3 numbers:", calc.multiply(2, 3, 4))

# Calling with more arguments
print("Multiplication of 4 numbers:", calc.multiply(1, 2, 3, 4))
```

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")


class Dog(Animal):
    def speak(self):
        print("Dog barks: Woof Woof")


class Cat(Animal):
    def speak(self):
        print("Cat meows: Meow Meow")


# Creating objects
a1 = Dog()
a2 = Cat()

# Polymorphism in action
animals = [a1, a2]

for animal in animals:
    animal.speak()
```

```python
import math

class Shape:

    @staticmethod
    def area(shape_type, *args):
        if shape_type == "circle":
            radius = args[0]
            return math.pi * radius * radius

        elif shape_type == "rectangle":
            length = args[0]
            width = args[1]
            return length * width

        else:
            return "Invalid shape type"


# Calling the static method
print("Area of Circle:", Shape.area("circle", 5))
print("Area of Rectangle:", Shape.area("rectangle", 4, 6))
```

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting...")


class Bike(Vehicle):
    def start(self):
        print("Bike starts with a self kick or button start.")


class Car(Vehicle):
    def start(self):
        print("Car starts with key ignition or push button.")


# Creating objects
v1 = Bike()
v2 = Car()

# Polymorphism in action
vehicles = [v1, v2]

for v in vehicles:
    v.start()
```

```python
class Printer:
    def print(self, data1=None, data2=None):
        
        # If only one argument is given
        if data1 is not None and data2 is None:
            print("Printing single value:", data1)
        
        # If both arguments are given
        elif data1 is not None and data2 is not None:
            print("Printing two values:", data1, data2)
        
        else:
            print("No data to print")


# Creating object
p = Printer()

# Different method calls (simulating overloading)
p.print("Hello World")       # string
p.print(100)                # integer
p.print("Age:", 20)        # string + integer
```

```python
class Person:
    def __init__(self, name):
        self.name = name


class Student(Person):
    def __init__(self, name, roll_no):
        super().__init__(name)
        self.roll_no = roll_no


# Checking subclass relationship
result = issubclass(Student, Person)

print("Is Student a subclass of Person?", result)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)


class Manager(Employee):
    def __init__(self, name, salary, department):
        # Calling parent class constructor using super()
        super().__init__(name, salary)
        self.department = department

    def display(self):
        super().display()
        print("Department:", self.department)


# Creating object
m1 = Manager("Amit", 50000, "IT")

# Display details
m1.display()
```

```python
class Grandparent:
    pass


class Parent(Grandparent):
    pass


class Child(Parent):
    pass


# Checking subclass relationships
print("Is Parent subclass of Grandparent?", issubclass(Parent, Grandparent))
print("Is Child subclass of Parent?", issubclass(Child, Parent))
print("Is Child subclass of Grandparent?", issubclass(Child, Grandparent))
print("Is Grandparent subclass of Child?", issubclass(Grandparent, Child))
```

```python
class Base:
    def display(self):
        print("This is the Base class method")


class Derived(Base):
    def display(self):
        # Calling base class method using super()
        super().display()
        print("This is the Derived class method")


# Creating object
obj = Derived()

# Calling method
obj.display()
```

```python
class User:
    def __init__(self, username, email):
        self.username = username
        self.email = email

    def display(self):
        print("Username:", self.username)
        print("Email:", self.email)


class Admin(User):
    def __init__(self, username, email, role):
        # Calling parent class constructor using super()
        super().__init__(username, email)
        self.role = role

    def display(self):
        super().display()
        print("Role:", self.role)


# Creating object
a1 = Admin("admin123", "admin@example.com", "Super Admin")

# Display details
a1.display()
```
