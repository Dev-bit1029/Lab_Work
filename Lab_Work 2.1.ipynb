```python
# take three integers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# find largest
if a >= b and a >= c:
    print("Largest number is:", a)
elif b >= a and b >= c:
    print("Largest number is:", b)
else:
    print("Largest number is:", c)
```

```python
# input distances of 3 cities
city1 = int(input("Enter distance of City A: "))
city2 = int(input("Enter distance of City B: "))
city3 = int(input("Enter distance of City C: "))

# find farthest city
if city1 > city2 and city1 > city3:
    print("City A is farthest")
elif city2 > city3:
    print("City B is farthest")
else:
    print("City C is farthest")
```
