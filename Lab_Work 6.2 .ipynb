```python
try:
    num1 = float(input("Enter numerator: "))
    num2 = float(input("Enter denominator: "))
    result = num1 / num2
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: You cannot divide by zero.")
except ValueError:
    print("Error: Please enter valid numeric values.")
```

```python
my_list = [10, 20, 30]
try:
    index = int(input("Enter index to access: "))
    print(f"Element at index {index}: {my_list[index]}")
except IndexError:
    print("Error: That index is out of range for this list.")
except ValueError:
    print("Error: Invalid input. Please enter an integer.")
```

```python
filename = input("Enter the filename to read: ")
try:
    with open(filename, 'r') as file:
        content = file.read()
except FileNotFoundError:
    print("Error: The file does not exist.")
except IOError:
    print("Error: Could not read the file.")
else:
    print("File read successfully! Content:")
    print(content)
```

```python
text = "Python"
try:
    idx = int(input("Enter string index: "))
    char = text[idx]
except IndexError:
    print("Error: Index out of bounds.")
except ValueError:
    print("Error: Please enter a whole number.")
else:
    print(f"The character at index {idx} is '{char}'.")
```

```python
try:
    f = open("sample.txt", "r")
    print(f.read())
except FileNotFoundError:
    print("Error: 'sample.txt' was not found.")
finally:
    # This ensures the file is closed if it was successfully opened
    if 'f' in locals() and not f.closed:
        f.close()
        print("File resources have been closed.")
```

```python
try:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    print(f"Result: {a / b}")
except (ValueError, ZeroDivisionError) as e:
    print(f"An error occurred: {e}")
finally:
    print("Division attempt finished. Thank you for using the program.")
```

```python
import math

try:
    val = float(input("Enter a number to find the square root: "))
    if val < 0:
        # Manually raising an error for negative numbers
        raise ValueError("Cannot calculate square root of a negative number.")
    res = math.sqrt(val)
except ValueError as e:
    print(f"Input Error: {e}")
else:
    print(f"The square root of {val} is {res:.2f}")
finally:
    print("Execution complete.")
```
