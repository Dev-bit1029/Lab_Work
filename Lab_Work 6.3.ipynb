```python
try:
    num = float(input("Enter a number: "))
    if num < 0:
        raise ValueError("Negative numbers are not allowed!")
    print(f"You entered: {num}")
except ValueError as e:
    print(f"Error: {e}")
```

```python
def check_even(val):
    if not isinstance(val, int):
        raise TypeError("Input must be an integer.")
    if val % 2 != 0:
        raise ValueError("The number is odd.")
    return f"{val} is even."

# Testing
try:
    print(check_even(10))  # Valid
    print(check_even(7))   # Raises ValueError
except (TypeError, ValueError) as e:
    print(f"Caught expected error: {e}")
```

```python
try:
    age = int(input("Enter your age: "))
    assert age > 18, "Access denied: Age must be above 18."
    print("Access granted.")
except AssertionError as e:
    print(e)
except ValueError:
    print("Please enter a valid numeric age.")
```

```python
def check_palindrome(s):
    assert len(s) > 0, "Input string cannot be empty."
    clean_s = s.replace(" ", "").lower()
    return clean_s == clean_s[::-1]

try:
    print(f"Is 'radar' a palindrome? {check_palindrome('radar')}")
    print(check_palindrome("")) # This will trigger the assertion
except AssertionError as e:
    print(f"Assertion Error: {e}")
```

```python
class InsufficientBalanceError(Exception):
    """Custom exception for bank withdrawals."""
    pass

def withdraw(amount, balance):
    if amount > balance:
        raise InsufficientBalanceError(f"Attempted to withdraw {amount}, but balance is only {balance}.")
    return balance - amount

# Testing
current_balance = 500
for test_amount in [200, 600]:
    try:
        current_balance = withdraw(test_amount, current_balance)
        print(f"Withdrawal successful! New balance: {current_balance}")
    except InsufficientBalanceError as e:
        print(f"Transaction Failed: {e}")
```

```python
class InvalidEmailError(Exception):
    pass

def validate_email(email):
    if "@" not in email or not (email.endswith(".com") or email.endswith(".org")):
        raise InvalidEmailError(f"'{email}' is not a valid email address.")
    return True

try:
    validate_email("user@example.net")
except InvalidEmailError as e:
    print(f"Validation Error: {e}")
```

```python
class InvalidGradeError(Exception):
    pass

def validate_grade(input_data):
    # Assert input is not empty
    assert input_data != "", "Grade input cannot be empty."
    
    grade = float(input_data)
    
    if grade < 0 or grade > 100:
        raise ValueError("Grade must be between 0 and 100.")
    if grade < 40:
        raise InvalidGradeError("Failing grade (below 40).")
    
    return f"Grade {grade} is valid."

try:
    print(validate_grade("35"))
except (ValueError, AssertionError, InvalidGradeError) as e:
    print(f"Error: {e}")
```

```python
class HighTemperatureError(Exception):
    pass

def process_temperature(temp):
    if not isinstance(temp, (int, float)):
        raise TypeError("Temperature must be a number.")
    
    # Check absolute range
    assert -273 <= temp <= 10000, "Temperature out of physical bounds (-273 to 10,000)."
    
    if temp > 1000:
        raise HighTemperatureError("Temperature exceeds 1,000°C; unrealistic for common use.")
    
    return (temp * 9/5) + 32  # Convert to Fahrenheit

try:
    print(f"Result: {process_temperature(1200)}°F")
except (TypeError, AssertionError, HighTemperatureError) as e:
    print(f"Safety Alert: {e}")
```
