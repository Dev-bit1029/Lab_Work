```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("-" * 20)


# Creating multiple objects
person1 = Person("Aarav", 20)
person2 = Person("Diya", 22)
person3 = Person("Rohan", 19)

# Calling method for each object
person1.display_details()
person2.display_details()
person3.display_details()
```

```python
class Counter:
    def __init__(self):
        self.count = 0   # initialized to zero

    def increment(self):
        self.count += 1  # increase count by 1

    def display(self):
        print("Current Count:", self.count)


# Creating object
c1 = Counter()

# Using methods
c1.display()     # 0
c1.increment()
c1.increment()
c1.display()     # 2
```

```python
class Book:
    def set_data(self, title, author):
        self.__title = title
        self.__author = author

    def get_data(self):
        print("Title:", self.__title)
        print("Author:", self.__author)


# Create object
b1 = Book()

# Set values
b1.set_data("Python Basics", "John Doe")

# Get values
b1.get_data()
```

```python
class Account:
    def set_balance(self, balance):
        self.__balance = balance   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def show(self):
        print("Balance:", self.__balance)


# Object
a1 = Account()

# Set starting balance
a1.set_balance(1000)

# Transactions
a1.deposit(500)
a1.withdraw(200)

# Show balance
a1.show()
```

```python
class Person:
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Invalid age")

    def get_age(self):
        return self.__age


# Create object
p1 = Person()

# Set age
p1.set_age(20)

# Get age
print("Age:", p1.get_age())
```

```python
class Student:
    def set_data(self, name, m1, m2, m3):
        self.__name = name
        self.__m1 = m1
        self.__m2 = m2
        self.__m3 = m3

    def calculate_average(self):
        self.avg = (self.__m1 + self.__m2 + self.__m3) / 3

    def display_average(self):
        print("Name:", self.__name)
        print("Average Marks:", self.avg)

    def display_grade(self):
        if self.avg >= 75:
            print("Grade: Distinction")
        elif self.avg >= 60:
            print("Grade: First Class")
        elif self.avg >= 40:
            print("Grade: Pass")
        else:
            print("Grade: Fail")


# Create object
s1 = Student()

# Set data
s1.set_data("Rahul", 80, 70, 90)

# Calculate and display
s1.calculate_average()
s1.display_average()
s1.display_grade()
```
