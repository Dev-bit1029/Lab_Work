```python
# Create a 3x3 matrix (2D list)
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Display the matrix in tabular format
print("3x3 Matrix:")
for row in matrix:
    for element in row:
        print(element, end="\t")  # tab space for alignment
    print()  # move to next line after each row
```

```python
# 2x3 matrix input
matrix = []

print("Enter 6 numbers (2x3 matrix):")

for i in range(2):
    row = []
    for j in range(3):
        row.append(int(input()))
    matrix.append(row)

# Transpose (3x2)
print("\nTranspose Matrix:")

for j in range(3):
    for i in range(2):
        print(matrix[i][j], end="\t")
    print()
```

```python
# Take input for 2x3 matrix
matrix = []

print("Enter elements for 2x3 matrix:")

for i in range(2):
    row = []
    for j in range(3):
        row.append(int(input()))
    matrix.append(row)

# Calculate sum of all elements
total = 0

for i in range(2):
    for j in range(3):
        total += matrix[i][j]

# Display result
print("\nSum of all elements:", total)
```

```python
# Create a list of integers
numbers = [5, 2, 9, 1, 7, 3]

# Sort the list in ascending order
numbers.sort()

# Display the sorted list
print("Sorted list (Ascending):", numbers)
```

```python
# List of tuples
data = [(1, 50), (3, 20), (2, 40), (4, 10)]

# Sort based on second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

# Display result
print("Sorted list based on 2nd element:")
print(sorted_data)
```

```python
# List of dictionaries
students = [
    {"name": "Amit", "marks": 75},
    {"name": "Ravi", "marks": 60},
    {"name": "Neha", "marks": 85},
    {"name": "Sita", "marks": 70}
]

# Sort by "marks"
sorted_students = sorted(students, key=lambda x: x["marks"])

# Display result
print("Sorted by marks:")
for student in sorted_students:
    print(student)
```

```python
# Sample list
numbers = [5, 2, 9, 1, 7]

# Using sort() -> modifies original list
list1 = numbers.copy()
list1.sort()
print("After sort() (in-place):")
print("list1:", list1)
print("Original list after sort():", numbers)

print("\n----------------------\n")

# Using sorted() -> creates new list
list2 = numbers.copy()
new_list = sorted(list2)
print("Using sorted() (new list):")
print("new_list:", new_list)
print("Original list after sorted():", list2)
```
