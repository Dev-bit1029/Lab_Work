```python
def factorial(n):
    return "Error" if n < 0 else 1 if n <= 1 else n * factorial(n - 1)
    
print(factorial(5))   
print(factorial(-3)) 
```

```python
def fibonacci(n):
    """
    Returns the nth Fibonacci number using recursion.

    Parameters:
    n (int): Position in Fibonacci sequence (n >= 0)

    Returns:
    int or str: Fibonacci number or error message
    """

    # Edge case: negative input
    if n < 0:
        return "Error: n must be non-negative"

    # Base cases
    elif n == 0:
        return 0
    elif n == 1:
        return 1

    # Recursive case
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


# Testing with various inputs
test_values = [0, 1, 2, 3, 5, 7, 10, -4]

for val in test_values:
    print(f"fibonacci({val}) =", fibonacci(val))
```

```python
def reverse(s):
    if s == "":
        return ""
    return reverse(s[1:]) + s[0]

# Test
print(reverse("hello"))   # olleh
print(reverse("python"))  # nohtyp
```

```python
def digit_sum(n):
    n = abs(n)  # handle negative numbers
    
    if n < 10:
        return n
    
    return digit_sum(n // 10 + n % 10)


# Test
print(digit_sum(9875))  # 2
print(digit_sum(1234))  # 1
```

```python
def is_prime(n, i=2):
    if n <= 1:
        return False
    if i == n:
        return True
    if n % i == 0:
        return False
    return is_prime(n, i + 1)


def primes(a, b):
    if a > b:
        return
    if is_prime(a):
        print(a, end=" ")
    primes(a + 1, b)


# Test
primes(10, 20)
```

```python

total = 0

def add_number(num):
    global total
    total += num
    print("Current sum:", total)


# Taking input
for _ in range(5): 
    n = int(input("Enter a number: "))
    add_number(n)

print("Final sum:", total)
```

```python
# Global variable
username = "Guest"

def update_username():
    global username
    new_name = input("Enter new username: ")
    username = new_name
    print("Username updated to:", username)


# Before update
print("Current username:", username)

# Update using function
update_username()

# After update
print("Final username:", username)
```

```python
def rectangle(l, w):
    area = l * w
    perimeter = 2 * (l + w)
    return area, perimeter

# Test
a, p = rectangle(5, 3)

print("Area:", a)
print("Perimeter:", p)
```

```python
def square_cube(n):
    return n*n, n*n*n   # returns (square, cube)

# Test
s, c = square_cube(3)

print("Square:", s)
print("Cube:", c)
```

```python
def split_string(s):
    vowels = "aeiouAEIOU"
    v = ""
    other = ""

    for ch in s:
        if ch in vowels:
            v += ch
        else:
            other += ch

    return v, other


# Test
a, b = split_string("Hello World")

print("Vowels:", a)
print("Others:", b)
```

```python
def split_words(words):
    vowels = "aeiouAEIOU"
    v_list = []
    c_list = []

    for w in words:
        if w[0] in vowels:
            v_list.append(w)
        else:
            c_list.append(w)

    return v_list, c_list


# Test
words = ["apple", "banana", "orange", "grape", "umbrella"]

v, c = split_words(words)

print("Vowel words:", v)
print("Consonant words:", c)
```
