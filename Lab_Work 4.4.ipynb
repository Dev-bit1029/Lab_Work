```python
# Take array size from user
n = int(input("Enter array size: "))

# Create empty arrays
a = []
b = []
c = []

print("Enter array A's elements:")

# Input for array A
for i in range(n):
    value = int(input(f"a[{i}] = "))
    a.append(value)

print("Enter array B's elements:")

# Input for array B
for i in range(n):
    value = int(input(f"b[{i}] = "))
    b.append(value)

# Perform addition
for i in range(n):
    sum_value = a[i] + b[i]
    c.append(sum_value)

# Output result
print("Array C is:", end=" ")

for i in range(n):
    if i == n - 1:
        print(c[i])
    else:
        print(c[i], end=", ")
```

```python
# Take array size
n = int(input("Enter array size: "))

# Create empty array
arr = []

print("Enter array elements:")

# Input elements
for i in range(n):
    value = int(input(f"a[{i}] = "))
    arr.append(value)

# Take number to search
key = int(input("Enter number to search: "))

# Search process
found = False

for i in range(n):
    if arr[i] == key:
        print("Element found at index:", i)
        found = True
        break

# If not found
if not found:
    print("Not Found")
```

```python
# Take array size
n = int(input("Enter array size: "))

# Create empty array
arr = []

print("Enter array elements:")

# Input elements
for i in range(n):
    value = int(input(f"a[{i}] = "))
    arr.append(value)

print("\nEven numbers in array are:")

# Print even numbers
for i in range(n):
    if arr[i] % 2 == 0:
        print(arr[i], end=" ")

print("\n\nOdd numbers in array are:")

# Print odd numbers
for i in range(n):
    if arr[i] % 2 != 0:
        print(arr[i], end=" ")
```

```python
##### Take array size
n = int(input("Enter array size: "))

# Create empty array
arr = []

print("Enter array elements:")

# Input elements
for i in range(n):
    value = int(input(f"a[{i}] = "))
    arr.append(value)

# Print first five elements
print("\nFirst five elements:")

limit = 5 if n >= 5 else n  # handle case if array has less than 5 elements

for i in range(limit):
    print(arr[i], end=" ")

# Print alternate elements
print("\n\nAlternate elements:")

for i in range(0, n, 2):
    print(arr[i], end=" ")
```

```python
# Take array size
n = int(input("Enter array size: "))

# Create empty array
arr = []

print("Enter array elements:")

# Input elements
for i in range(n):
    value = int(input(f"a[{i}] = "))
    arr.append(value)

# First element
print("\nFirst element:", arr[0])

# Last element
print("Last element:", arr[n - 1])

# Middle element(s)
print("Middle element(s):")

if n % 2 == 1:
    # If odd size → one middle element
    mid = n // 2
    print(arr[mid])
else:
    # If even size → two middle elements
    mid1 = (n // 2) - 1
    mid2 = n // 2
    print(arr[mid1], arr[mid2])
```
