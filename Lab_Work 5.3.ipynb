```python
# Parent class
class Parent:
    def display(self):
        print("This is the Parent class method")

# Child class inheriting Parent
class Child(Parent):
    def show(self):
        print("This is the Child class method")

# Creating object of Child class
obj = Child()

# Calling method from Parent class
obj.display()

# Calling method from Child class
obj.show()
```

```python
# First parent class
class Teacher:
    def teach(self):
        print("Teacher is teaching students.")

# Second parent class
class Administrator:
    def manage(self):
        print("Administrator is managing school activities.")

# Child class inheriting both Teacher and Administrator
class Headmaster(Teacher, Administrator):
    def lead(self):
        print("Headmaster is leading the school.")

# Creating object of Headmaster class
obj = Headmaster()

# Calling methods from both parent classes and child class
obj.teach()
obj.manage()
obj.lead()
```

```python
# Grandparent class
class Grandparent:
    def show_grandparent(self):
        print("I am the Grandparent")

# Parent class inheriting Grandparent
class Parent(Grandparent):
    def show_parent(self):
        print("I am the Parent")

# Child class inheriting Parent
class Child(Parent):
    def show_child(self):
        print("I am the Child")

# Creating object of Child class
obj = Child()

# Calling methods from all levels
obj.show_grandparent()
obj.show_parent()
obj.show_child()
```

```python
# Base class
class Animal:
    def show_type(self):
        print("I am an Animal")

# First subclass
class Dog(Animal):
    def bark(self):
        print("Dog barks: Woof Woof!")

# Second subclass
class Cat(Animal):
    def meow(self):
        print("Cat meows: Meow Meow!")

# Creating objects
dog_obj = Dog()
cat_obj = Cat()

# Calling methods from base class and subclasses
print("Dog object:")
dog_obj.show_type()
dog_obj.bark()

print("\nCat object:")
cat_obj.show_type()
cat_obj.meow()
```

```python
# Base class
class Person:
    def show(self):
        print("I am a Person")

# Multilevel Inheritance
class Employee(Person):
    def show(self):
        super().show()   # Calls Person class method
        print("I am an Employee")

# Another base class for multiple inheritance
class Manager:
    def show(self):
        print("I am a Manager")

# Hybrid Inheritance: Employee + Manager
class TeamLead(Employee, Manager):
    def show(self):
        super().show()   # Resolves ambiguity using MRO (Method Resolution Order)
        print("I am a Team Lead")

# Creating object
obj = TeamLead()

# Calling method
obj.show()
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

# Creating an object
obj = Student("Dev", 20)

# Using dir() to list attributes and methods
print("Attributes and Methods of Student class/object:")
print(dir(obj))
```

```python
# Creating a class
class Student:
    def __init__(self, name):
        self.name = name

# Creating objects
obj1 = Student("Dev")
obj2 = "Hello"

# Checking if objects are instances of Student class
print("obj1 is instance of Student:", isinstance(obj1, Student))
print("obj2 is instance of Student:", isinstance(obj2, Student))
```
